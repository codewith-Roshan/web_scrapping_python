```python
for i in soup.find_all('h2'): #to access only h2 heading in webpage
  print(i.text.strip()) #text strip hepl get only text value in heading
```

```python
company = soup.find_all('div', class_="mt-0 mb-std p-std css-1ax1pfu css-errlgf") #here company has div and class name so we access all the div in single webpage
```

here loop running for find paragraph in div and text strip give desired value

```python
for i in soup.find('p', class_="css-1sj9xzx css-56kyx5"):
  print(i.text.strip())
```

creating empty list and with the help of loop we append their values in the list

```python
name = []
rating = []
reviews = []
salaries =[]
jobs = []
location = []
global_company_size = []
industry = []
desc = []
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find_all('span', class_="pr-xsm ratingsWidget__RatingsWidgetStyles__rating")[0].text.strip())
  reviews.append(i.find('div', class_="d-flex flex-column align-items-center").text.strip())
  salaries.append(i.find_all('h3', class_="mt-xsm mt-md-0")[1].text.strip())
  jobs.append(i.find_all('h3', class_="mt-xsm mt-md-0")[2].text.strip())
  location.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[0].text.strip())
  global_company_size.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[1].text.strip())
  industry.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[2].text.strip())
  try:
        desc.append(i.find('p', class_="css-1sj9xzx css-56kyx5").text.strip())
  except:
      desc.append(np.nan)


```

```python
d = {'Name':name, 'Rating':rating, 'Reviews':reviews, 'Salaries': salaries, 'Jobs':jobs, 'Loction':location, 'Global_Company_Size':global_company_size, 'Industry':industry, 'Description':desc}

df = pd.DataFrame(d)
df
```

here the final code with accessing all pages in the website with the help of j loop then left code is same as above

```python
final = pd.DataFrame()
for j in range(1,701):
  
  #formate method use to auto increment the page in url
  headers = {'user-Agent': 'Mozilla/5.0 (Window NT 6.3; Win64; x64) AppleWebKit/537.36(KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get('https://www.glassdoor.co.in/Reviews/index.htm?overall_rating_low=3.5&page={}&filterType=RATING_OVERALL'.format(j), headers=headers).text
  soup = BeautifulSoup(webpage, 'lxml') #html parsing

  company = soup.find_all('div', class_="mt-0 mb-std p-std css-1ax1pfu css-errlgf")
  name = []
  rating = []
  reviews = []
  salaries =[]
  jobs = []
  location = []
  global_company_size = []
  industry = []
  desc = []
  for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find_all('span', class_="pr-xsm ratingsWidget__RatingsWidgetStyles__rating")[0].text.strip())
    reviews.append(i.find('div', class_="d-flex flex-column align-items-center").text.strip())
    salaries.append(i.find_all('h3', class_="mt-xsm mt-md-0")[1].text.strip())
    jobs.append(i.find_all('h3', class_="mt-xsm mt-md-0")[2].text.strip())
    location.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[0].text.strip())
    global_company_size.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[1].text.strip())
    industry.append(i.find_all('span', class_="d-block mt-0 css-56kyx5")[2].text.strip())
    try:
        desc.append(i.find('p', class_="css-1sj9xzx css-56kyx5").text.strip())
    except:
      desc.append(np.nan)

  d = {'Name':name, 'Rating':rating, 'Reviews':reviews, 'Salaries': salaries, 'Jobs':jobs, 'Loction':location, 'Global_Company_Size':global_company_size, 'Industry':industry, 'Description':desc}
  df = pd.DataFrame(d)
  final =final.append(df,ignore_index=True)




```
